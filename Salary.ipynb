### Loop to find number of unique values to identify the irrelevant independent variables

```python
for c in dataset_tr.columns:
    print ("---- %s ---" % c)
    print (dataset_tr[c].nunique())
```

### Deleting irrelevant columns

```python
dataset_te.drop(['age','educationno','capitalgain','capitalloss','hoursperweek','native'],axis=1,inplace=True)
dataset_tr.drop(['age','educationno','capitalgain','capitalloss','hoursperweek','native'],axis=1,inplace=True)
```

```python
dataset_tr
```

### Separate dependent and independent variable

```python
x_train=dataset_tr.iloc[:,:-1]
y_train=dataset_tr.iloc[:,-1]
x_test=dataset_te.iloc[:,:-1]
y_test=dataset_te.iloc[:,-1]
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

### Converting independent variable to 0,1,2.....

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
string_columns=x_train.columns
for i in string_columns:
    x_train[i]=le.fit_transform(x_train[i])
    x_train[i]=le.fit_transform(x_train[i])
```

```python
string_columns1=x_test.columns
for i in string_columns1:
    x_test[i]=le.fit_transform(x_test[i])
    x_test[i]=le.fit_transform(x_test[i])
```

```python
x_test
```

### Encoding Dependent Variable

```python
y_train = le.fit_transform(y_train)
y_test = le.fit_transform(y_test)
```

```python
y_train
```

```python
y_test
```

### Model Building

```python
from sklearn.naive_bayes import GaussianNB
Gmodel=GaussianNB()
```

```python
Gmodel.fit(x_train,y_train)
```

```python
y_pred=Gmodel.predict(x_test)
```

```python
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
